```python
import joblib

#load the data 
x_train = joblib.load('x_train.joblib')
x_test = joblib.load('x_test.joblib')
y_train = joblib.load('y_train.joblib')
y_test = joblib.load('y_test.joblib')
x_train_resampled = joblib.load('x_train_resampled.joblib')
y_train_resampled = joblib.load('y_train_resampled.joblib')
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,precision_score,recall_score
```

```python
import mlflow
import mlflow.sklearn
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("SMS Spam Classification")
```

```python
with mlflow.start_run(run_name="xbgrid grid search smote"):
    parameters = {'n_estimators': [100, 200, 300], 'learning_rate': [0.1, 0.01, 0.001]}
    grid_xb_smote = GridSearchCV(xgb.XGBClassifier(), parameters, cv=5, scoring='accuracy')
    grid_xb_smote.fit(x_train_resampled, y_train_resampled)
    y_pred_xb_smote = grid_xb_smote.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_xb_smote)
    f1 = f1_score(y_test,y_pred_xb_smote)
    recall = recall_score(y_test,y_pred_xb_smote)
    precision = precision_score(y_test,y_pred_xb_smote)
    print("XGBoost with SMOTE:")
    print(f"Accuracy: {accuracy_score(y_test,y_pred_xb_smote)}")
    print(f"F1 Score: {f1_score(y_test,y_pred_xb_smote)}")
    print(f"Recall: {recall_score(y_test,y_pred_xb_smote)}")
    print(f"Precision: {precision_score(y_test,y_pred_xb_smote)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test,y_pred_xb_smote)}")
    print(f"Classification Report:\n{classification_report(y_test,y_pred_xb_smote)}")
    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)    
    mlflow.log_param("n_estimators", grid_xb_smote.best_params_['n_estimators'])
    mlflow.log_param("learning_rate", grid_xb_smote.best_params_['learning_rate'])
    mlflow.sklearn.log_model(grid_xb_smote.best_estimator_, name="xb_model_grid_search_smote")
    
```
