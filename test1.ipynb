```python
import joblib

#load the data 
x_train = joblib.load('x_train.joblib')
x_test = joblib.load('x_test.joblib')
y_train = joblib.load('y_train.joblib')
y_test = joblib.load('y_test.joblib')
x_train_resampled = joblib.load('x_train_resampled.joblib')
y_train_resampled = joblib.load('y_train_resampled.joblib')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,precision_score,recall_score
```

```python
import mlflow
import mlflow.sklearn
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("SMS Spam Classification")
```

```python
with mlflow.start_run(run_name="Logistic Regression Model class_weight balanced"):
    lr_model = LogisticRegression(class_weight='balanced')
    lr_model.fit(x_train,y_train)
    y_pred_lr = lr_model.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_lr)
    f1 = f1_score(y_test,y_pred_lr)
    recall = recall_score(y_test,y_pred_lr)
    precision = precision_score(y_test,y_pred_lr)
    print("Logistic Regression:")
    print(accuracy_score(y_test,y_pred_lr))
    print(f1_score(y_test,y_pred_lr))
    print(recall_score(y_test,y_pred_lr))
    print(precision_score(y_test,y_pred_lr))
    print(confusion_matrix(y_test,y_pred_lr))
    print(classification_report(y_test,y_pred_lr))

    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)

    mlflow.log_param("class_weight", "balanced")

    mlflow.sklearn.log_model(lr_model, name="lr_model_balanced")
```

```python
# grid search 

from sklearn.model_selection import GridSearchCV

with mlflow.start_run(run_name="Logistic Regression Grid Search"):
    lr = LogisticRegression()
    parameters = {'C': [0.1, 1, 10]}
    grid_lr = GridSearchCV(lr, parameters, cv=5, scoring='accuracy')
    grid_lr.fit(x_train,y_train)
    y_pred_lr = grid_lr.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_lr)
    f1 = f1_score(y_test,y_pred_lr)
    recall = recall_score(y_test,y_pred_lr)
    precision = precision_score(y_test,y_pred_lr)
    print("Logistic Regression:")
    print(accuracy_score(y_test,y_pred_lr))
    print(f1_score(y_test,y_pred_lr))
    print(recall_score(y_test,y_pred_lr))
    print(precision_score(y_test,y_pred_lr))
    print(confusion_matrix(y_test,y_pred_lr))
    print(classification_report(y_test,y_pred_lr))
    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.log_param("C", grid_lr.best_params_['C'])
    mlflow.sklearn.log_model(grid_lr.best_estimator_, name="lr_model_grid_search")
```

```python
with mlflow.start_run(run_name="logistic regression smote"):
    lr_smote = LogisticRegression()
    lr_smote.fit(x_train_resampled, y_train_resampled)
    y_pred_lr_smote = lr_smote.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_lr_smote)
    f1 = f1_score(y_test,y_pred_lr_smote)
    recall = recall_score(y_test,y_pred_lr_smote)
    precision = precision_score(y_test,y_pred_lr_smote)
    print("Logistic Regression with SMOTE:")
    print(accuracy_score(y_test,y_pred_lr_smote))
    print(f1_score(y_test,y_pred_lr_smote))
    print(recall_score(y_test,y_pred_lr_smote))
    print(precision_score(y_test,y_pred_lr_smote))
    print(confusion_matrix(y_test,y_pred_lr_smote))
    print(classification_report(y_test,y_pred_lr_smote))

    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)

    mlflow.sklearn.log_model(lr_smote, name="lr_model_smote")
```

```python
# grid search with smote

with mlflow.start_run(run_name="Logistic Regression Grid Search Smote"):
    parameters = {'C': [0.1, 1, 10]}
    grid_lr_smote = GridSearchCV(LogisticRegression(), parameters, cv=5, scoring='accuracy')
    grid_lr_smote.fit(x_train_resampled, y_train_resampled)
    y_pred_lr_smote = grid_lr_smote.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_lr_smote)
    f1 = f1_score(y_test,y_pred_lr_smote)
    recall = recall_score(y_test,y_pred_lr_smote)
    precision = precision_score(y_test,y_pred_lr_smote)
    print("Logistic Regression with SMOTE:")
    print(accuracy_score(y_test,y_pred_lr_smote))
    print(f1_score(y_test,y_pred_lr_smote))
    print(recall_score(y_test,y_pred_lr_smote))
    print(precision_score(y_test,y_pred_lr_smote))
    print(confusion_matrix(y_test,y_pred_lr_smote))
    print(classification_report(y_test,y_pred_lr_smote))
    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.log_param("C", grid_lr_smote.best_params_['C'])
    mlflow.sklearn.log_model(grid_lr_smote.best_estimator_, name="lr_model_grid_search_smote")
```
