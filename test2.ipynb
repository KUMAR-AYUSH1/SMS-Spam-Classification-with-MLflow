```python
import joblib

#load the data 
x_train = joblib.load('x_train.joblib')
x_test = joblib.load('x_test.joblib')
y_train = joblib.load('y_train.joblib')
y_test = joblib.load('y_test.joblib')
x_train_resampled = joblib.load('x_train_resampled.joblib')
y_train_resampled = joblib.load('y_train_resampled.joblib')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,precision_score,recall_score
```

```python
import mlflow
import mlflow.sklearn
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("SMS Spam Classification")
```

```python
with mlflow.start_run(run_name="naive bayes smote"):
    nb_smote = GaussianNB()
    nb_smote.fit(x_train_resampled.toarray(),y_train_resampled)
    y_pred_nb_smote = nb_smote.predict(x_test.toarray())
    accuracy = accuracy_score(y_test,y_pred_nb_smote)
    f1 = f1_score(y_test,y_pred_nb_smote)
    recall = recall_score(y_test,y_pred_nb_smote)
    precision = precision_score(y_test,y_pred_nb_smote)
    print("Naive Bayes with SMOTE:")
    print(accuracy_score(y_test,y_pred_nb_smote))
    print(f1_score(y_test,y_pred_nb_smote))
    print(recall_score(y_test,y_pred_nb_smote))
    print(precision_score(y_test,y_pred_nb_smote))
    print(confusion_matrix(y_test,y_pred_nb_smote))
    print(classification_report(y_test,y_pred_nb_smote))

    mlflow.log_metric("accuracy",accuracy)  
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)

    mlflow.sklearn.log_model(nb_smote, name="nb_model_smote")
```

```python
with mlflow.start_run(run_name="naive bayes grid search smote"):
    nb = GaussianNB()
    parameters = {'var_smoothing': [1e-09, 1e-08, 1e-07]}
    grid_nb_smote = GridSearchCV(nb, parameters, cv=5, scoring='accuracy')
    grid_nb_smote.fit(x_train_resampled.toarray(), y_train_resampled)
    y_pred_nb_smote = grid_nb_smote.predict(x_test.toarray())
    accuracy = accuracy_score(y_test,y_pred_nb_smote)
    f1 = f1_score(y_test,y_pred_nb_smote)
    recall = recall_score(y_test,y_pred_nb_smote)
    precision = precision_score(y_test,y_pred_nb_smote)
    print("Naive Bayes with SMOTE:")
    print(accuracy_score(y_test,y_pred_nb_smote))
    print(f1_score(y_test,y_pred_nb_smote))
    print(recall_score(y_test,y_pred_nb_smote))
    print(precision_score(y_test,y_pred_nb_smote))
    print(confusion_matrix(y_test,y_pred_nb_smote))
    print(classification_report(y_test,y_pred_nb_smote))
    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)  
    mlflow.log_metric("precision",precision)
    mlflow.log_param("var_smoothing", grid_nb_smote.best_params_['var_smoothing'])
    mlflow.sklearn.log_model(grid_nb_smote.best_estimator_, name="nb_model_grid_search_smote")
    
```
