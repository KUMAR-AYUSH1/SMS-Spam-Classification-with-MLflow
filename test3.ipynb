```python
import joblib

#load the data 
x_train = joblib.load('x_train.joblib')
x_test = joblib.load('x_test.joblib')
y_train = joblib.load('y_train.joblib')
y_test = joblib.load('y_test.joblib')
x_train_resampled = joblib.load('x_train_resampled.joblib')
y_train_resampled = joblib.load('y_train_resampled.joblib')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,precision_score,recall_score
```

```python
import mlflow
import mlflow.sklearn
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("SMS Spam Classification")
```

```python
with mlflow.start_run(run_name="SVC Model class_weight balanced"):

    svc_model = SVC(class_weight='balanced')
    svc_model.fit(x_train,y_train)
    y_pred_svc = svc_model.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_svc)
    f1 = f1_score(y_test,y_pred_svc)
    recall = recall_score(y_test,y_pred_svc)
    precision = precision_score(y_test,y_pred_svc)
    print("SVC:")
    print(accuracy_score(y_test,y_pred_svc))
    print(f1_score(y_test,y_pred_svc))
    print(recall_score(y_test,y_pred_svc))
    print(precision_score(y_test,y_pred_svc))
    print(confusion_matrix(y_test,y_pred_svc))
    print(classification_report(y_test,y_pred_svc))
    mlflow.log_metric("accuracy",accuracy)
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.log_param("class_weight", 'balanced')
    mlflow.sklearn.log_model(svc_model, name="svc_model_balanced")
```

```python
with mlflow.start_run(run_name="SVC Grid Search"):
    svc = SVC()
    parameters = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
    grid_svc = GridSearchCV(svc, parameters, cv=5, scoring='accuracy')
    grid_svc.fit(x_train,y_train)
    y_pred_svc = grid_svc.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_svc)
    f1 = f1_score(y_test,y_pred_svc)
    recall = recall_score(y_test,y_pred_svc)
    precision = precision_score(y_test,y_pred_svc)
    print("SVC:")
    print(accuracy_score(y_test,y_pred_svc))
    print(f1_score(y_test,y_pred_svc))
    print(recall_score(y_test,y_pred_svc))
    print(precision_score(y_test,y_pred_svc))
    print(confusion_matrix(y_test,y_pred_svc))
    print(classification_report(y_test,y_pred_svc))
    mlflow.log_metric("accuracy",accuracy)  
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.log_param("C", grid_svc.best_params_['C'])
    mlflow.log_param("kernel", grid_svc.best_params_['kernel'])
    mlflow.sklearn.log_model(grid_svc.best_estimator_, name="svc_model_grid_search")    
    
```

```python
with mlflow.start_run(run_name="SVC smote"):
    svc_smote = SVC()
    svc_smote.fit(x_train_resampled,y_train_resampled)
    y_pred_svc_smote = svc_smote.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_svc_smote)
    f1 = f1_score(y_test,y_pred_svc_smote)
    recall = recall_score(y_test,y_pred_svc_smote)
    precision = precision_score(y_test,y_pred_svc_smote)
    print("SVC with SMOTE:")
    print(accuracy_score(y_test,y_pred_svc_smote))
    print(f1_score(y_test,y_pred_svc_smote))
    print(recall_score(y_test,y_pred_svc_smote))
    print(precision_score(y_test,y_pred_svc_smote))
    print(confusion_matrix(y_test,y_pred_svc_smote))
    print(classification_report(y_test,y_pred_svc_smote))
    mlflow.log_metric("accuracy",accuracy)  
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.sklearn.log_model(svc_smote, name="svc_model_smote")
```

```python
with mlflow.start_run(run_name="SVC Grid Search smote"):
    parameters = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
    grid_svc_smote = GridSearchCV(SVC(), parameters, cv=5, scoring='accuracy')
    grid_svc_smote.fit(x_train_resampled, y_train_resampled)
    y_pred_svc_smote = grid_svc_smote.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred_svc_smote)
    f1 = f1_score(y_test,y_pred_svc_smote)
    recall = recall_score(y_test,y_pred_svc_smote)
    precision = precision_score(y_test,y_pred_svc_smote)
    print("SVC with SMOTE:")
    print(accuracy_score(y_test,y_pred_svc_smote))
    print(f1_score(y_test,y_pred_svc_smote))
    print(recall_score(y_test,y_pred_svc_smote))
    print(precision_score(y_test,y_pred_svc_smote))
    print(confusion_matrix(y_test,y_pred_svc_smote))
    print(classification_report(y_test,y_pred_svc_smote))
    mlflow.log_metric("accuracy",accuracy)  
    mlflow.log_metric("f1",f1)
    mlflow.log_metric("recall",recall)
    mlflow.log_metric("precision",precision)
    mlflow.log_param("C", grid_svc_smote.best_params_['C'])
    mlflow.log_param("kernel", grid_svc_smote.best_params_['kernel'])
    mlflow.sklearn.log_model(grid_svc_smote.best_estimator_, name="svc_model_grid_search_smote")
```
